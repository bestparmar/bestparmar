```python
def Flip_dataframe(df,df2):
    for i in df.urldb:
        url = i
        response = requests.get(url)
        soup = BeautifulSoup(response.content,'html.parser')
        try:
            Item_name =soup.find('span',{'class':'B_NuCI'}).text
        except:
            pass
        try:
            Product_Specifications = soup.find('div',{'class':'_2cM9lP'}).text
            Product_Highlights = Product_Specifications.split("Highlights")
            Product_Highlights = Product_Highlights[1]
        except:
            pass
        try:
            global Product_description
            Product_description = soup.find('div',{'class':'_1mXcCf RmoJUa'}).text
        except:
            pass
        attr=soup.find_all('td',class_='_1hKmbr col col-3-12')
        valu=soup.find_all('td',class_='URwL2w col col-9-12')
        
        
        values=[i.text for i in valu]

        df2 = df2.append({"url":i,"Item_name":Item_name,"Product_Highlights":Product_Highlights,"Product_description":Product_description,"Specifications":dict(zip(attributes,values))})
    return df2
```
